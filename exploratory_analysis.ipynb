```python
import pandas as pd
import os.path
import matplotlib.pyplot as plt
import numpy as np
import zipfile
import wget
```

```python
# can i extract zip directly from lahman??.  do existence checks plus override (delete) in final version
lahman_zip_url = 'https://github.com/chadwickbureau/baseballdatabank/archive/refs/tags/v2022.2.zip'
lahman_zip_file = "./Resources/lahman.zip"

wget.download(url=lahman_zip_url, out=lahman_zip_file)
```

```python
zf = zipfile.ZipFile(lahman_zip_file)
# teams_csv = os.path.join("Resources", "baseballdatabank-master", "core", "Teams.csv")

teams_csv = zf.open('baseballdatabank-2022.2/core/Teams.csv')

teams_df = pd.read_csv(teams_csv)
teams_df.head(5)
```
