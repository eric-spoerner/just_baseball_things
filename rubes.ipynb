```python
year_bins = np.arange(1875,1931,step=5.0)
group_names = [
    "1876-1880",
    "1881-1885",
    "1886-1890",
    "1891-1895",
    "1896-1900",
    "1901-1905",
    "1906-1910",    
    "1911-1915",
    "1916-1920",
    "1921-1925",
    "1926-1930"
]

# rubes_range = 
rubes_bins = pd.cut(x=rubes_birthyear, bins=year_bins, labels=group_names) #.groupby(year_bins).count()
rubes_cnt = rubes_bins.groupby(rubes_bins).count()

#how to rotate the x labels?
plt.bar(x=rubes_cnt.index
        ,height=rubes_cnt)

plt.xticks(rotation='vertical')
plt.title("when rubes born??")
plt.xlabel("year")
plt.ylabel("numrubes")

plt.show()
```
